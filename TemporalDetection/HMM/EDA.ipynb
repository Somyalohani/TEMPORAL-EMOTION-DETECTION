```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = open("data1.txt", "r")
pdata = file.read();
data = pdata.encode("utf-8").decode("utf-8").split("\n")
```

```python
lis1=[]
lis2=[]
lis3=[]
lis4=[]
for dic in data:
    try:
        dicti = eval(dic)
        lis1.append(dicti['measurements']['interocularDistance'])
        lis2.append(dicti['measurements']['orientation']['pitch'])
        lis3.append(dicti['measurements']['orientation']['yaw'])
#     lis4.append(dicti['measurements']['orientation']['roll'])
    except:
        SyntaxError
```

```python
df = pd.DataFrame()
df['InterocularDistance'] = lis1
df['Pitch'] = lis2
df['Yaw'] = lis3
# df['Roll'] = lis4
```

```python
sns.lineplot(df.index,df['Pitch'])
plt.show
```
